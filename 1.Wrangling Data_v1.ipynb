### Custom aggregations

#### Calculate product wise  no. days between purchases

```python
df
```

```python
list(df.groupby(['product_cat']))
```

```python
for index, group in df.groupby(['product_cat']):
    print(index)
```

```python
for index, group in df.groupby(['product_cat']):
    display(group)
```

```python
for index, group in df.groupby(['product_cat']):
    print('Group: ',index)
    print('Total Sales:',group.sales.sum())
    display(group)
    print('----')
```

```python
pd.Series([10,20,30,40]).diff()
```

```python
for index, group in df.groupby(['product_cat']):
    print(index)
    group = group.sort_values('sale_date')
    group['dates_diff'] = group['sale_date'].diff().dt.days
    display(group)
```

#### Calculate product wise  no. Months between purchases???

```python
for index, group in df.groupby(['product_cat']):
    print(index)
    group = group.sort_values('sale_date')
    group['month_diff'] = group['sale_date'].dt.month.diff()
    display(group)
```
